# Important Libraries

```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

# Loading the Dataset 

```python
df=pd.read_csv('/Users/tuhinalambijoy/Downloads/data.csv')
df
```

Droping the id colum

```python
df=df.drop(columns='id')
```

Cheaking is there any null values or not

```python
df.isnull().sum()
```

We can see that there is a colum name is Unnamed : 32 , and that has 569 null values , so removing the colum

```python
df=df.drop(columns='Unnamed: 32')
```

Then again cheaking is there any null value left or not

```python
df.isnull().sum()
```

Cheaking is there any duplicates or not

```python
df.drop_duplicates()
```

We can see there is no colum deleted so that's mean there is no duplicates

# Feature 

Separating the the feature and target 

```python
X=df.drop(columns='diagnosis')
y=df['diagnosis']
```

```python
X
```

```python
y
```

# Scaling the feature 

```python
from sklearn.preprocessing import MinMaxScaler
mms=MinMaxScaler()
X_scale=mms.fit_transform(X)
```

```python
X_scale
```

# Spliting into train and test

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X_scale,y,test_size=0.2,random_state=42)
```

# Model Training

```python
from sklearn.svm import SVC
svm_model=SVC(kernel='rbf',gamma='auto',C=5)
svm_model.fit(X_train,y_train)
```

# Cheaking the training data accuracy

```python
svm_model.score(X_train,y_train)
```

# Test data accuracy

```python
svm_model.score(X_test,y_test)
```

# Cross cheak validation

```python
from sklearn.model_selection import cross_val_score
```

```python
svm1=cross_val_score(SVC(kernel='linear',gamma='auto',C=5),X_scale,y,cv=5)
svm1
```

```python
svm1.mean()
```

```python
svm2=cross_val_score(SVC(kernel='rbf',gamma='auto',C=5),X_scale,y,cv=5)
svm2
```

```python
svm2.mean()
```

## Hyper Parameter tuning ##

```python
from sklearn.model_selection import GridSearchCV
```

```python
Hyper_Para={
    'gamma':['auto','scale'],
    'C':[1,10,20,30],
    'kernel':['rbf','linear','poly','sigmoid']
}

clf=GridSearchCV(SVC(),Hyper_Para,cv=5)
```

```python
clf.fit(X_train,y_train)
```

```python
clf.cv_results_
```

```python
new_df=pd.DataFrame(clf.cv_results_)
new_df
```

# Finding the best Parameters

```python
clf.best_params_
```

# Best score using those parameter

```python
clf.best_score_
```

So We found that if we use C=1 , gamma=scale and kernel=rbf  we can get the best accuracy

Now cheaking the accuracy

```python
svm_model=SVC(kernel='rbf',gamma='scale',C=1)
svm_model.fit(X_train,y_train)
```

```python
svm_model.score(X_train,y_train)
```

```python
svm_model.score(X_test,y_test)
```
